# Ablation Study / Backward Feature Elimination on final drop 2

```python
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import root_mean_squared_error
from xgboost import XGBRegressor

def train_eval(train_df, dev_df, features):
    model = XGBRegressor(
        n_estimators=400,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )

    model.fit(train_df[features], train_df["aqipm25"])
    preds = model.predict(dev_df[features])

    rmse = root_mean_squared_error(
        dev_df["aqipm25"], preds
    )
    return rmse
```

```python
all_features = train_df.drop(columns=["aqipm25","date"]).columns.tolist()
baseline_rmse = train_eval(train_df, dev_df, all_features)
baseline_rmse
```

```python
ablation_results = {}

for f in final_drop_2:
    feats = [x for x in all_features if x != f]
    rmse = train_eval(train_df, dev_df, feats)
    ablation_results[f] = rmse - baseline_rmse
```

```python
BASE_FEATURES = [
    c for c in train_df.columns
    if c not in ["aqipm25", "date"]
]
```

```python
REDUCED_FEATURES = [
    c for c in BASE_FEATURES
    if c not in test_drop_features
]
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import root_mean_squared_error,r2_score

def train_eval_rmse(train_df, dev_df, features, seed=42):
    model = XGBRegressor(
        n_estimators=400,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=seed,
        n_jobs=-1
    )

    model.fit(train_df[features], train_df["aqipm25"])
    preds = model.predict(dev_df[features])

    rmse = root_mean_squared_error(
        dev_df["aqipm25"],
        preds
    )
    r2=r2_score(dev_df['aqipm25'],preds)
    return rmse,r2
```

```python
rmse_full,r2_full = train_eval_rmse(
    train_df,
    dev_df,
    BASE_FEATURES
)

rmse_reduced,r2_reduced = train_eval_rmse(
    train_df,
    dev_df,
    REDUCED_FEATURES
)

rmse_full, rmse_reduced
```

```python
r2_full,r2_reduced
```

```python
delta = rmse_reduced - rmse_full
delta
```

```python
seeds = [0, 21, 42, 99]
results = []

for s in seeds:
    full,full1 = train_eval_rmse(train_df, dev_df, BASE_FEATURES, seed=s)
    reduced,reduced1 = train_eval_rmse(train_df, dev_df, REDUCED_FEATURES, seed=s)
    results.append({
        "seed": s,
        "rmse_full": full,
        "rmse_reduced": reduced,
        "delta": reduced - full,
        "r2-full":full1,
        "r2-reduced":reduced1,
        "delt2": reduced1 - full1,
    })

pd.DataFrame(results)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(data):
    data = data.dropna()
    vif_df = pd.DataFrame({
        "Column": data.columns,
        "VIF": [
            variance_inflation_factor(data.values, i)
            for i in range(data.shape[1])
        ]
    })
    return vif_df.sort_values("VIF", ascending=False)
```

# pca tested

```python
TARGET = "aqipm25"
ID_COLS = ["date"]
```

```python
X_train = train_df.drop(columns=[TARGET] + ID_COLS)
y_train = train_df[TARGET]

X_dev = dev_df.drop(columns=[TARGET] + ID_COLS)
y_dev = dev_df[TARGET]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_dev_scaled   = scaler.transform(X_dev)

```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.9999, random_state=42)

X_train_pca = pca.fit_transform(X_train_scaled)
X_dev_pca   = pca.transform(X_dev_scaled)


print("Original features:", X_train.shape[1])
print("PCA components :", X_train_pca.shape[1])
```

```python
from xgboost import XGBRegressor

model_pca = XGBRegressor(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model_pca.fit(X_train_pca, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score

def evaluate(model, X, y, name=""):
    preds = model.predict(X)
    print(f"\n{name}")
    print("RMSE:", root_mean_squared_error(y, preds))
    print("MAE :", mean_absolute_error(y, preds))
    print("R2  :", r2_score(y, preds))

evaluate(model_pca, X_dev_pca, y_dev, "DEV (PCA)")

```
