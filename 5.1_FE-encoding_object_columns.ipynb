```python
import pandas as pd
```

```python
df=pd.read_csv('data/data2_nafilled.csv')
df.info()
```

# 1)

```python
df['preciptype'].value_counts()
```

```python
df["preciptype"] = df["preciptype"].map({
    "rain": 1,
    "no rain": 0
}).astype(int)
```

```python
df['preciptype'].value_counts()
```

 # 2)

```python
df['conditions'].value_counts()
```

```python
df.drop(columns=["conditions"], inplace=True)
```

# 3)

```python
df['description'].value_counts()
```

```python
df.drop(columns=["description"], inplace=True)
```

# 4)

```python
df['icon'].value_counts()
```

```python
df = pd.get_dummies(
    df,
    columns=["icon"],
    prefix="icon",
    drop_first=False
)
```

```python
df.info()
```

```python
df.drop(columns=["icon_wind"], inplace=True)
```

```python
df.filter(like="icon_").sum()
```

```python
df.info()
```

```python
bool_cols = df.select_dtypes(include="bool").columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.info()
```

# 5)

```python
df["sunrise"] = pd.to_datetime(df["sunrise"])
df["sunset"]  = pd.to_datetime(df["sunset"])
df["date"] = pd.to_datetime(df["date"])
```

```python
df["daylight_minutes"] = (df["sunset"] - df["sunrise"]).dt.total_seconds() / 60
```

```python
df.head(3)
```

```python
df.drop(["sunrise", "sunset"], axis=1, inplace=True)
```

# 6)

```python
df.info()
```

```python
df['aqipm25']=df['aqi']
```

```python
df.drop(['pm25','aqi_source','aqi'],axis=1,inplace=True)
```

```python
df.info()
```

```python
datetime_cols = df.select_dtypes(include=["datetime64[ns]"]).columns.tolist()
int_cols      = df.select_dtypes(include=["int64"]).columns.tolist()
float_cols    = df.select_dtypes(include=["float64"]).columns.tolist()
```

```python
df = df[datetime_cols + int_cols + float_cols]
```

```python
df.info()
```

```python
df.to_csv('data/data3_encoded5.csv',index=False)
```
